```python
import pandas as pd
import requests
import sqlite3
from io import StringIO
url = 'https://www.football-data.co.uk/mmz4281/1920/E0.csv'
response = requests.get(url).text
df = pd.read_csv(StringIO(response))
df.head()
```

```python
date_time = []
for d , t  in zip(df['Date'],df['Time']):
    date_time.append(f'{d}:{t}')
```

```python
div       = df['Div']
home_team = df['HomeTeam']
away_team  = df['AwayTeam']
fthg      = df['FTHG']
ftag      = df['FTAG']
```

```python
conn = sqlite3.connect('footballTable.db')
c = conn.cursor()
def create_table():
    c.execute("""CREATE TABLE IF NOT EXISTS football(id INTEGER AUTOINCREMENT ,
                            div VARCHAR NOT NULL,
                            date_time TIMESTAMP NOT NULL,
                            home_team VARCHAR NOT NULL,
                            away_team VARCHAR NOT NULL,
                            fthg INT DEFAULT(0),
                            ftag INT DEFAULT(0),
                          CONSTRAINT rid_pkey PRIMARY KEY (id))""")
```
